```python
# 3. String Methods: Write a Python program that takes a string, strM = “Python beginner tutorial” and perform the following tasks:
# a. Convert the sentence to uppercase.
# b. Convert the sentence to lowercase.
# c. Use Capitalize and return the sentence to the original input form.
# d. Count the total number of occurrences of character ‘t’  in the string.
# e. Replace all occurrences of “Python” with “Machine Learning”  in the input string strM = “Python beginner tutorial”

strM = "Python beginner tutorial"

print("a. Sentence to uppercase: " + strM.upper())
print("b. Sentence to lowercase: " + strM.lower())
print("c. Capitalize and retuning sentence to original : " + strM.capitalize()) # For captializing each word we use title() Ex. strM.title()
print("d. Count of total no. of occurrences char 't': " , strM.count('t'))
print("e. Replacing Python with Machine Learning : " + strM.replace('Python', 'Machine Learning'))

```
