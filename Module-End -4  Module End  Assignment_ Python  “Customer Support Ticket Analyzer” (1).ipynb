# Add More Tickets

```python
#Ask the user:  How many new tickets do you want to add?
#For each new ticket, collect:
#Customer Name
# Issue Description
# Priority (High / Medium / Low)

#Requirements:
#Ticket numbers must auto-increment starting from 11.
#Validate priority (accept only High, Medium, Low).
#Append new data to ticket_data.

Adding_Ticket = int(input("How many new tickets do you want to add?"))

next_ticket_no = max(ticket_data['Ticket_No']) + 1

for i in range(Adding_Ticket):
    customer_name = input("Enter Customer Name : ")
    issue_desc = input("Enter Issue Description : ")

    while True:
        priority = input("Priority (High / Medium / Low)")
        if priority in ['High', 'Medium', 'Low']:
            break
        print("Invalid priority! Please enter High, Medium, or Low.")

    ticket_data['Ticket_No'].append(next_ticket_no)
    ticket_data['Customer_Name'].append(customer_name)
    ticket_data['Issue_Description'].append(issue_desc)
    ticket_data['Priority'].append(priority)

    next_ticket_no += 1

print(f"Added {Adding_Ticket} new tickets. Total tickets: {len(ticket_data['Ticket_No'])}")

print("\nCUSTOMER SUPPORT TICKET DATA")
for i in range(len(ticket_data['Ticket_No'])):
    print(f"Ticket {nums[i]:2d} | Name: {names[i]:7s} | Priority: {prios[i]:7s} | Issue: {issues[i]}")
```

# Text Cleaning for Issue Descriptions

```python
#Clean all issue descriptions by applying:
#Remove punctuation (.,!?-)
# Convert multiple spaces → single space
# Remove leading/trailing spaces
# Convert text to lowercase
# Replace slang/shorthand (example: "ok" → "okay")
def clean_text(text):

    text = text.replace('ok', 'okay') #Replacing Slang
    text = text.replace('ur', 'your')
    text = text.replace('u', 'you')

    punctuations = "!@#$%^&*.,!?-"

    for punct in punctuations:
        text = text.replace(punct, ' ') #Replacing Puncations

    text = ' '.join(text.split()) # multiple spaces → single space
    text = text.strip().lower() #removes Leading/trailing spaces and converts to lower

    return text

print("\nCleaning all issue descriptions")
for i in range(len(ticket_data['Issue_Description'])):
    ticket_data['Issue_Description'][i] = clean_text(ticket_data['Issue_Description'][i])

print("\nCleaning Completed")
```

# Keyword-Based Issue Insights

```python
#Create a function:
#def count_tickets_with_word(word):
#Function Requirements:
#Case-insensitive search
#Returns how many ticket descriptions contain the given word

def count_tickets_with_word(word):
  word = word.lower().strip()
  count = 0
  for description in ticket_data['Issue_Description']:
    if word in description.lower():
      count += 1
  return count

print(f"Tickets containing 'poor' is: {count_tickets_with_word('poor')}")
print(f"Tickets containing 'good' is: {count_tickets_with_word('good')}")
print(f"Tickets containing 'slow' is: {count_tickets_with_word('slow')}")
print(f"Tickets containing 'excellent' is: {count_tickets_with_word('excellent')}")
```

```python
# 2. Priority Analysis
# Number of High-priority tickets
# Number of Medium priority tickets
# Number of Low-Priority tickets

print("\nPRIORITY ANALYSIS")
high_count = ticket_data['Priority'].count('High')
medium_count = ticket_data['Priority'].count('Medium')
low_count = ticket_data['Priority'].count('Low')

print(f"High:    {high_count}")
print(f"Medium:  {medium_count}")
print(f"Low:     {low_count}")
```

```python
# 3. Find the Ticket With the Longest Issue Description
# Based on word count, print:
# Ticket number
# Customer name
# Cleaned issue text
# Word count

longest_words = 0
longest_ticket_no = 0
longest_name = ""
longest_issue = ""

for i in range(len(ticket_data['Issue_Description'])):
    word_count = len(ticket_data['Issue_Description'][i].split())

    if word_count > longest_words:
        longest_words = word_count
        longest_ticket_no = ticket_data['Ticket_No'][i]
        longest_name = ticket_data['Customer_Name'][i]
        longest_issue = ticket_data['Issue_Description'][i]

print("\nLongest Issue Description:")
print(f"Ticket {longest_ticket_no}")
print(f"Name : {longest_name}")
print(f"Issue: {longest_issue}")
print(f"Words: {longest_words}")
```

```python
# 4. Extract Unique Words Used
# Generate a set of all unique words across all issue descriptions.
# Show:
# Count of unique words
# The word list (sorted)

all_words = []

for issue in ticket_data['Issue_Description']:
    words = issue.split()
    all_words = all_words + words

word_set = set(all_words)
unique_words = sorted(word_set)

print("Unique Words Used")
print("Count of unique words:", len(unique_words))
print("All unique words:", unique_words)
```
